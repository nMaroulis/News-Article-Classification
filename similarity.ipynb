```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
import csv
import io, os, errno
import fileinput
import collections as cl
from os import path
import nltk
# nltk.download("stopwords")
from nltk.corpus import stopwords 
from sklearn.metrics.pairwise import  cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# stop_words = set(stopwords.words('english'))


dup_csv = "./datasets/duplicatePairs.csv"

try:
    os.remove(dup_csv)
except OSError:
    pass

trainset_file = "./datasets/train_set.csv"
train_df = pd.read_csv(trainset_file,  sep='\t')

sim_threshold = 0.7
train_df.groupby('Category').size()
```

Calculate the Matrix containing the word terms using Tf-Idf. Then apply cosine similarity between all of them.

```python
similarity_df = train_df['Content'].head(np.size(train_df,0)).values

tfidf_vectorizer = TfidfVectorizer()

tfidf_matrix = tfidf_vectorizer.fit_transform(similarity_df)

cos_similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

# max_similarity = [0]*tfidf_matrix.shape[0]
similarity_vector = []

for i in range(tfidf_matrix.shape[0]):
    for c in range(len(cos_similarity_matrix[i])):
        if cos_similarity_matrix[i][c] > sim_threshold  and c != i:
            if([c,i,cos_similarity_matrix[i][c]] not in similarity_vector): # prevent duplicate entry
                similarity_vector.append([i,c,cos_similarity_matrix[i][c]])
    if i % 1000:
        print('Checked ', i)

# print(similarity_vector)
```

Append results to duplicatePairs.csv

```python
with open(dup_csv, mode='w') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=['Document_ID1', 'Document_ID2', 'Similarity'], delimiter='\t')
    writer.writeheader()

    for sim in similarity_vector:
        writer.writerow({'Document_ID1': sim[0], 'Document_ID2': sim[1], 'Similarity': sim[2]})
```
